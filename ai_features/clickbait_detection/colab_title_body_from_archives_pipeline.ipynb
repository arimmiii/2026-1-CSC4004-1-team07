# Clickbait Shared Data Prep
`clickbait_detection/data/{train,valid,test}.csv`를 Colab 로컬로 복사하고 구조를 확인하는 노트북입니다.
학습은 별도 노트북에서 진행합니다.


```python
from google.colab import drive
drive.mount('/content/drive')

import os
import shutil
import pandas as pd
```

```python
DRIVE_ROOT = '/content/drive/MyDrive/sw_project/ai_features/clickbait_detection'
DRIVE_DATA_DIR = os.path.join(DRIVE_ROOT, 'data')
LOCAL_DATA_DIR = '/content/clickbait_detection_data'

assert os.path.exists(DRIVE_DATA_DIR), f'Not found: {DRIVE_DATA_DIR}'

if os.path.exists(LOCAL_DATA_DIR):
    shutil.rmtree(LOCAL_DATA_DIR)
shutil.copytree(DRIVE_DATA_DIR, LOCAL_DATA_DIR)

print('copied to:', LOCAL_DATA_DIR)
print('files:', sorted(os.listdir(LOCAL_DATA_DIR)))
```

```python
for name in ['train.csv', 'valid.csv', 'test.csv']:
    path = os.path.join(LOCAL_DATA_DIR, name)
    df = pd.read_csv(path)
    print('\n', name, 'rows=', len(df), 'cols=', list(df.columns))
    print(df.head(2).to_string(index=False))
```
